### CSV faylni yangilash

Quyidagi kod namunasi CSV faylni o'qish, undagi ma'lumotlarni o'zgartirish va o'zgarishlarni yangi CSV fayliga saqlash jarayonini ko'rsatadi.

Birinchi qadamda `pandas` kutubxonasini import qilamiz va mavjud CSV faylni (masalan, `/tmp/data.csv`) `DataFrame` ob'ektiga yuklaymiz. Agar sizda bu fayl bo'lmasa, test uchun bitta yaratishingiz mumkin.

```python
import pandas as pd

# Namuna CSV fayl yaratish (agar sizda mavjud bo'lmasa)
# Bu qismni real faylingiz bo'lsa o'tkazib yuborishingiz mumkin.
data = {'Column1': [10, 20, 30, 40, 50],
        'Column2': ['A', 'B', 'C', 'D', 'E'],
        'Column3': [100, 200, 300, 400, 500]}
sample_df = pd.DataFrame(data)
sample_df.to_csv('/tmp/data.csv', index=False)

# CSV faylni o'qish
try:
    df = pd.read_csv('/tmp/data.csv')
    print("Fayl muvaffaqiyatli yuklandi:")
    display(df.head())
except FileNotFoundError:
    print("Xato: '/tmp/data.csv' fayli topilmadi. Yuqoridagi namuna faylni yaratish kodini ishga tushiring yoki to'g'ri fayl yo'lini kiriting.")
```

Endi `DataFrame` ustidagi ma'lumotlarni o'zgartiramiz. Misol uchun, `Column1` ustunining har bir qiymatiga 10 qo'shamiz.

```python
# Ma'lumotlarni o'zgartirish. Misol: 'Column1' ustuniga 10 qo'shish
# O'zingizning ehtiyojlaringizga qarab bu qismni o'zgartirishingiz mumkin.
if 'Column1' in df.columns:
    df['Column1'] = df['Column1'] + 10
    print("O'zgartirilgan DataFrame:")
    display(df.head())
else:
    print("Xato: 'Column1' ustuni topilmadi. O'zgartirmoqchi bo'lgan ustun nomini tekshiring.")
```

O'zgartirilgan `DataFrame`ni yangi CSV fayliga saqlaymiz. Original faylni saqlashni istasangiz, avval uning nusxasini yaratishni unutmang.

```python
# O'zgartirilgan DataFrame'ni yangi CSV fayliga saqlash
output_file_path = '/tmp/updated_data.csv'
df.to_csv(output_file_path, index=False)

print(f"O'zgartirilgan ma'lumotlar '{output_file_path}' fayliga saqlandi.")

# Saqlangan faylni tekshirish (ixtiyoriy)
# updated_df = pd.read_csv(output_file_path)
# display(updated_df.head())
```
