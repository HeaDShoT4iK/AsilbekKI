### Pipeline jarayonlari va Datasetni tayyorlash

`id` va `Unnamed: 32` (agar mavjud bo'lsa) kabi keraksiz ustunlarni olib tashlaymiz, `diagnosis` ustunini raqamli qiymatlarga (0 va 1) kodlaymiz, so'ngra ma'lumotlarni o'qitish va test to'plamlariga ajratamiz hamda xususiyatlarni standartlashtiramiz.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np

# 'id' ustunini tashlab yuboramiz, chunki u model uchun foydasiz
# 'Unnamed: 32' ustunini ham tashlab yuboramiz, chunki u bo'sh
df_processed = df_uploaded.copy()

if 'id' in df_processed.columns:
    df_processed = df_processed.drop('id', axis=1)
if 'Unnamed: 32' in df_processed.columns:
    df_processed = df_processed.drop('Unnamed: 32', axis=1)

# Maqsad ustunini (target variable) 'diagnosis' sifatida aniqlaymiz
X = df_processed.drop('diagnosis', axis=1) # X - xususiyatlar (features)
y = df_processed['diagnosis'] # y - maqsad (target)

# 'M' (Malignant) va 'B' (Benign) qiymatlarini raqamli ko'rinishga keltiramiz (1 va 0)
y = y.map({'M': 1, 'B': 0})

print('Maʼlumotlar tayyorlandi. Xususiyatlar (X) va maqsad (y) aniqlandi.')

# Ma'lumotlarni o'qitish va test to'plamlariga bo'lish
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Xususiyatlarni standartlashtirish (K-NN uchun muhim)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print('\nMaʼlumotlar oʻqitish va test toʻplamlariga boʻlindi va standartlashtirildi.')
print(f'Oʻqitish toʻplami hajmi: {X_train_scaled.shape}')
print(f'Test toʻplami hajmi: {X_test_scaled.shape}')

# Scale qilingan ma'lumotlarni DataFrame'ga qaytarish (agar kerak bo'lsa, vizualizatsiya uchun)
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X.columns)

print('\nX_train_scaled ning dastlabki 5 qatori:')
display(X_train_scaled_df.head())
```

### K-NN modelini o'qitish va natijalarini baholash

K-NN modelini turli `K` (qo'shnilar soni) qiymatlari uchun o'qitamiz va har bir `K` uchun modelning aniqligini (accuracy) hisoblaymiz. Keyin natijalarni grafikda ko'rsatamiz.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

accuracy_scores = []
k_values = range(1, 31) # K qiymatlari 1 dan 30 gacha

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    y_pred = knn.predict(X_test_scaled)
    accuracy_scores.append(accuracy_score(y_test, y_pred))

# Grafik chizish
plt.figure(figsize=(12, 6))
plt.plot(k_values, accuracy_scores, marker='o', linestyle='-', color='skyblue')
plt.title('K-NN aniqligi K qiymatiga qarab')
plt.xlabel('K qiymati (Neighbors soni)')
plt.ylabel('Aniqlik (Accuracy)')
plt.xticks(list(k_values))
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Eng yaxshi K qiymatini topish
best_k = k_values[np.argmax(accuracy_scores)]
max_accuracy = np.max(accuracy_scores)

print(f'Eng yuqori aniqlik {max_accuracy:.4f} bilan {best_k} K qiymatida erishildi.')

# Eng yaxshi K qiymati bilan modelni qayta o'qitish va to'liq hisobotni ko'rish
print(f'\nEng yaxshi K={best_k} uchun model baholash:')
best_knn = KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(X_train_scaled, y_train)
y_pred_best_k = best_knn.predict(X_test_scaled)

print(f'Test toʻplamidagi aniqlik: {accuracy_score(y_test, y_pred_best_k):.4f}')
print('\nKlassifikatsiya hisoboti:')
print(classification_report(y_test, y_pred_best_k))
```
