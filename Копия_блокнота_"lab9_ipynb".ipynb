```python
import sqlite3
import pandas as pd

# Define the name of your database file
db_name = 'sentences.db'

# Connect to SQLite database (it will be created if it doesn't exist)
conn = sqlite3.connect(db_name)
cursor = conn.cursor()

# Create a table to store sentences
# Each row will have an ID and the sentence text
cursor.execute('''
    CREATE TABLE IF NOT EXISTS sentences (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        text TEXT NOT NULL
    )
''')
conn.commit()

print(f"Database '{db_name}' created and 'sentences' table ensured.")
```

```python
from google.colab import files
import io
import pandas as pd

uploaded = files.upload()

for fn in uploaded.keys():
    print('User uploaded file "{name}" with length {length} bytes'.format(
        name=fn, length=len(uploaded[fn])))

    # Read the content of the uploaded file as text lines
    file_content_str = uploaded[fn].decode('utf-8')
    sentences = file_content_str.splitlines() # Split by lines

    # Insert each sentence into the database
    # Ensure 'conn' and 'cursor' are defined from the previous cell
    if 'conn' in globals() and 'cursor' in globals():
        for sentence in sentences:
            clean_sentence = sentence.strip()
            if clean_sentence:
                cursor.execute("INSERT INTO sentences (text) VALUES (?) ", (clean_sentence,))
        conn.commit()
        print(f"Successfully loaded {len(sentences)} lines from '{fn}' into the database.")

        # Display first 5 sentences from the database to verify
        print("\nFirst 5 sentences in the database:")
        df = pd.read_sql_query("SELECT * FROM sentences LIMIT 5", conn)
        display(df)
    else:
        print("Error: SQLite database connection (conn, cursor) not found. Please run the SQLite setup cell first.")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import sqlite3

# Connect to the database
conn = sqlite3.connect(db_name)

# Read sentences from the database
df_sentences = pd.read_sql_query("SELECT text FROM sentences", conn)
sentences_list = df_sentences['text'].tolist()

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(
    max_df=0.85, # ignore terms that appear in more than 85% of documents
    min_df=5,    # ignore terms that appear in less than 5 documents
    stop_words=None # you might want to add Uzbek stop words if available
)

# Fit and transform the sentences to get TF-IDF vectors
X = tfidf_vectorizer.fit_transform(sentences_list)

print(f"Sentences vectorized into a matrix of shape: {X.shape}")
print("TF-IDF vectorization complete. The vectors are stored in variable 'X'.")
```
