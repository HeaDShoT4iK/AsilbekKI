```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import pandas as pd

# Datasetni yuklash
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normallashtirish (0-255 oraliqni 0-1 oraliqqa o'tkazamiz)
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
def create_model(activation_func):
    model = models.Sequential([
        layers.Flatten(input_shape=(28, 28)),          # Input layer
        layers.Dense(128, activation=activation_func), # Hidden layer 1
        layers.Dense(64, activation=activation_func),  # Hidden layer 2
        layers.Dense(10, activation='softmax')         # Output layer (10 ta raqam uchun)
    ])

    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

activations = ['sigmoid', 'relu', 'tanh']
history_dict = {}

# Har bir funksiya uchun modelni o'qitish
for func in activations:
    print(f"\nO'qitilmoqda: {func} funksiyasi bilan...")
    model = create_model(func)
    history = model.fit(x_train, y_train, epochs=10, validation_split=0.2, verbose=0)
    history_dict[func] = history
```

```python
plt.figure(figsize=(12, 5))

# Accuracy grafigi
plt.subplot(1, 2, 1)
for func in activations:
    plt.plot(history_dict[func].history['val_accuracy'], label=func)
plt.title('Validation Accuracy')
plt.legend()

# Loss grafigi
plt.subplot(1, 2, 2)
for func in activations:
    plt.plot(history_dict[func].history['val_loss'], label=func)
plt.title('Validation Loss')
plt.legend()

plt.show()
```
