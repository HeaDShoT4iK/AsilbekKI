#### 2. Statistik ma'lumotlarni olish va saqlash

DataFrame yuklangan bo'lsa, endi statistik funksiyalarni ishlatamiz va natijalarni alohida CSV fayllariga saqlaymiz.

```python
if df is not None:
    # df.describe() - deskriptiv statistikani hisoblash
    desc_stats = df.describe()
    print("\nDescriptive Statistics:")
    display(desc_stats)
    desc_stats.to_csv('describe.csv')
    print("Descriptive statistics 'describe.csv' fayliga saqlandi.")
else:
    print("DataFrame yuklanmaganligi sababli statistik hisob-kitoblar o'tkazilmadi.")
```

```python
if df is not None:
    # df.corr() - juftlik korrelyatsiyasini hisoblash (faqat raqamli ustunlar uchun)
    # Raqamli ustunlarni tanlab olamiz
    numeric_df = df.select_dtypes(include=['number'])
    if not numeric_df.empty:
        correlation_matrix = numeric_df.corr()
        print("\nCorrelation Matrix:")
        display(correlation_matrix)
        correlation_matrix.to_csv('corr.csv')
        print("Correlation matrix 'corr.csv' fayliga saqlandi.")
    else:
        print("DataFrame'da raqamli ustunlar topilmadi, shuning uchun korrelyatsiya hisoblanmadi.")
else:
    print("DataFrame yuklanmaganligi sababli statistik hisob-kitoblar o'tkazilmadi.")
```

```python
if df is not None:
    # df.cov() - kovariatsiyani hisoblash (faqat raqamli ustunlar uchun)
    # Raqamli ustunlarni tanlab olamiz
    numeric_df = df.select_dtypes(include=['number'])
    if not numeric_df.empty:
        covariance_matrix = numeric_df.cov()
        print("\nCovariance Matrix:")
        display(covariance_matrix)
        covariance_matrix.to_csv('cov.csv')
        print("Covariance matrix 'cov.csv' fayliga saqlandi.")
    else:
        print("DataFrame'da raqamli ustunlar topilmadi, shuning uchun kovariatsiya hisoblanmadi.")
else:
    print("DataFrame yuklanmaganligi sababli statistik hisob-kitoblar o'tkazilmadi.")
```
