```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 1. Rasmga o'xshash tasodifiy ma'lumotlar yaratamiz
# n_samples - nuqtalar soni, centers - klasterlar soni
X, y = make_blobs(n_samples=3000, centers=3, cluster_std=0.7, random_state=42)

# 2. KMeans modelini sozlaymiz
# k=3 (n_clusters), init="k-means++" parametri bilan
k_means = KMeans(init="k-means++", n_clusters=3, n_init=12)

# 3. Modelni o'qitish
k_means.fit(X)

# 4. Natijani vizuallash (Rasmda ko'rsatilganidek grafik chizish)
plt.figure(figsize=(8, 6))

# Nuqtalarni klasterlarga qarab bo'yash (labels_)
plt.scatter(X[:, 0], X[:, 1], c=k_means.labels_, marker='.', s=10, cmap='plasma', alpha=0.5)

# Klaster markazlarini (sentroidlarni) ko'rsatish
centers = k_means.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=50, alpha=0.9, marker='o', edgecolors='white')

plt.title("KMeans klasterlash natijasi (k=3)")
plt.xticks([]) # X o'qi raqamlarini o'chirish (rasmdagidek)
plt.yticks([]) # Y o'qi raqamlarini o'chirish
plt.show()
```
