### 1. Ma'lumotlarni yuklash va tozalash

```python
import pandas as pd
import numpy as np

# Datasetni yuklash (fayl nomi 'mushrooms.csv' deb faraz qilindi)
try:
    df = pd.read_csv('mushrooms.csv')
    print("Ma'lumotlar muvaffaqiyatli yuklandi.")
except FileNotFoundError:
    print("Xato: 'mushrooms.csv' fayli topilmadi. Iltimos, fayl nomini tekshiring yoki to'g'ri joylashtiring.")
    # Agar fayl topilmasa, namunaviy DataFrame yaratish
    data = {
        'class': np.random.choice(['e', 'p'], size=100),
        'cap-shape': np.random.choice(['x', 'b', 's', 'f', 'k', 'c'], size=100),
        'cap-surface': np.random.choice(['f', 'g', 'y', 's'], size=100),
        'cap-color': np.random.choice(['n', 'b', 'c', 'g', 'r', 'p', 'u', 'e', 'w', 'y'], size=100),
        'odor': np.random.choice(['a', 'l', 'c', 'y', 'f', 'm', 'n', 'p', 's'], size=100),
        'gill-attachment': np.random.choice(['a', 'f'], size=100),
        'gill-spacing': np.random.choice(['c', 'w', 'd'], size=100),
        'gill-size': np.random.choice(['b', 'n'], size=100),
        'gill-color': np.random.choice(['k', 'n', 'b', 'h', 'g', 'r', 'o', 'p', 'u', 'e', 'w', 'y'], size=100)
    }
    df = pd.DataFrame(data)
    print("Namunaviy DataFrame yaratildi.")

# Ma'lumotlarning dastlabki 5 qatorini ko'rsatish
display(df.head())

# Ma'lumotlar turini va null qiymatlarini tekshirish
print("\nMa'lumotlar haqida qisqacha ma'lumot:")
df.info()
```

```python
# Null qiymatlarni tekshirish va tozalash
print("\nNull qiymatlarni tekshirish:")
print(df.isnull().sum())

# Agar '?' qiymatlari mavjud bo'lsa (ba'zi datasetlarda bo'lishi mumkin)
# Ularni NaN deb o'zgartirish va keyin o'chirish
df.replace('?', np.nan, inplace=True)
print("\n'?' qiymatlari NaN ga o'zgartirilgandan keyin null qiymatlar:")
print(df.isnull().sum())

# NaN (null) qiymatlarni o'chirish (agar mavjud bo'lsa)
df.dropna(inplace=True)
print(f"\nNaN qiymatlari o'chirilgandan so'ng qatorlar soni: {len(df)}")
```

```python
# Kategorik ustunlarni aniqlash va one-hot encoding qilish
# 'class' ustuni maqsad ustun deb hisoblanadi

# Maqsad ustunini ajratish
y = df['class']
X = df.drop('class', axis=1)

# Kategorik ustunlarni one-hot encoding qilish
X = pd.get_dummies(X, drop_first=True)

# Maqsad ustunini (y) Label Encoding qilish (e=0, p=1)
y = y.map({'e': 0, 'p': 1})

print("\nOne-hot encodingdan keyingi X:")
display(X.head())
print("\nLabel encodingdan keyingi y:")
display(y.head())
```
